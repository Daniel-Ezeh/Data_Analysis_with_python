## 3.	Figure and Axes:
You can use the `figure` function to resize and specify the resolution of your plot.

### How to Add Axes to a Figure
We can add axes to a figure in matplotlib by passing a list argument in the `add_axes()` method.

Syntax: matplotlib.pyplot.figure.add_axes(rect)

rect: This parameter is the dimensions `[xmin, ymin, dx, dy]` of the new axes. It takes the below elements as arguments in the list:

- `xmin`: Horizontal coordinate of the lower left corner.
- `ymin`: Vertical coordinate of the lower left corner.
- `dx`: Width of the subplot.
- `dy`: Height of the subplot.

Returns: This method return the axes class depends on the projection used. 

```python
# Example
# Creating two axis
import numpy as np
import matplotlib.pyplot as plt

# Create figure() objects 
# This acts as a container  
# for the different plots 
fig = plt.figure(figsize=(4, 4))

# Generating a line graph
x = np.arange(0,1.414*2, 0.05)
y1 = np.sin(x)
y2 = np.cos(x)


# Creating two axes
#add_axes([xmin,ymin,dx,dy]) 
ax1 = fig.add_axes([0,0,1,1])
ax1.plot(x,y1, )

ax2 = fig.add_axes([0,1,1,0.5])
ax2.plot(x,y2)

plt.show()
```

### Set Font Properties for Title and Labels
You can use the fontdict parameter in `xlabel()`, `ylabel()`, and `title()` to set font properties for the title and labels.

```python
y1 = np.array([127.08,  26.,   148.,   100.,    10.69, 127.5 ])
y2 = np.array([301.32,  28.,   479.,   119.,    73.62, 291.2 ])

font1 = {'family':'serif', 'color':'blue', 'size':20, "fontweight":"bold"}
font2 = {'family':'Tahoma', 'color':'darkred', 'size':14}

plt.plot(y1, "o-y", ms=20, mec="#4CAF50", mfc="red")
plt.plot(y2, "*-r", ms=15, mec="blue", mfc="hotpink")

plt.xlabel("Serialize", fontdict=font2)
plt.ylabel("Maxpulse and Pulse", fontdict=font2)

plt.title("A Graph of Pulse and Maxpulse", fontdict=font1)

plt.show()
```

#### Annotating text
The uses of the basic `text` function above place text at an arbitrary position on the Axes.

A common use for text is to annotate some feature of the plot, and the `annotate` method provides helper functionality to make annotations easy. 

In an annotation, there are two points to consider: the location being annotated represented by the argument `xy` and the location of the text `xytext`. Both of these arguments are `(x, y)` tuples.

```python
y1 = np.array([127.08,  26.,   148.,   100.,    10.69, 127.5 ])
y2 = np.array([301.32,  28.,   479.,   119.,    73.62, 291.2 ])

font1 = {'family':'serif', 'color':'blue', 'size':20}
font2 = {'family':'Tahoma', 'color':'darkred', 'size':14}

line1 = plt.plot(y1, "D--r", ms=5, mec="black", mfc="darkblue", alpha=0.3)
line2 = plt.plot(y2, "*-k", ms=8, mec="red", mfc="hotpink")


# The tuple uses the x and y axis scale to place the annotation and arrowhead
plt.annotate("Maximum point", xy=(2,480), xytext=(3,485),
            arrowprops=dict(facecolor="black", shrink=0.05)
            )

plt.title("Pulse and Maxpulse", fontdict=font1, loc="center")



plt.show()
```
