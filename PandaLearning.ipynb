```python
  import numpy as np
from scipy.stats import ks_2samp
# Data for Class A and Class B exam scores
class_a_scores = np.array([78, 82, 85, 90, 88, 92, 95, 89, 86, 91])
class_b_scores = np.array([76, 80, 82, 85, 84, 89, 86, 83, 79, 80])

# Perform the K-S test
ks_statistic, p_value = ks_2samp(class_a_scores, class_b_scores)

# Define the significance level (alpha)
alpha = 0.05

# Output the results
print("K-S Statistic:", ks_statistic)
print("P-value:", p_value)

if p_value > alpha:
    print("Fail to reject the null hypothesis: The two samples come from the same distribution.")
else:
    print("Reject the null hypothesis: The two samples come from different distributions.")
```

#### 2. Reading From SQL: pd.read_sql()

In pandas, the pd.read_sql() function is used to read data from a SQL database into a DataFrame. This function allows you to execute SQL queries and retrieve the results directly into a DataFrame. Here's how to use it:


Here's a breakdown of the steps:

Import the pandas library.

Import the appropriate database connector, such as sqlite3, and create a connection to your database. You'll need to replace 'your_database.db' with the path to your database or provide the necessary connection details for your database system.

Write your SQL query and store it in the sql_query variable. This can be any valid SQL query that retrieves data from your database.

Use pd.read_sql(sql_query, connection) to execute the SQL query and load the results into a DataFrame (df in this example).

Close the database connection when you're done with it.

Now, you can work with the data stored in the DataFrame as you would with any other pandas DataFrame. This approach allows you to read data from a SQL database and perform various data analysis and manipulation tasks using pandas.

```python

import pandas as pd
import sqlite3  # You can use the appropriate database connector for your database system

# Create a connection to your database
connection = sqlite3.connect('your_database.db')  # Replace with your database connection details

# Write your SQL query
sql_query = "SELECT * FROM your_table"

# Use pd.read_sql to execute the query and load the results into a DataFrame
df = pd.read_sql(sql_query, connection)

# Close the database connection when you're done
connection.close()

# Now, you can work with the data in the DataFrame
print(df)
```

If you need to work with different database systems, such as MySQL, PostgreSQL, Oracle, SQL Server, or others, you will need to use the respective database connectors or libraries that are designed for those systems. Here are some popular Python libraries for different databases:

- MySQL: You can use the mysql-connector-python or pymysql library to work with MySQL databases.
- PostgreSQL: The psycopg2 library is commonly used to connect to PostgreSQL databases.
- Oracle: For Oracle databases, the cx_Oracle library is the recommended choice.
- SQL Server: The pyodbc- library is a popular option for connecting to Microsoft SQL Server.
- NoSQL Databases: For NoSQL databases like MongoDB, you can use libraries like pymongo.

- Data in wrong format
- Wrong data

- `id_vars` : A list of columns that you want to keep as-is in the long format. In this example, we keep the "Name" column as-is.
