# COMBINING DATA

```python
# Creating a DataFrame Object

a = {
    "S/N": [1,2,3,4,5,6,7,8],
    "Colors": ["Blue", "Black", "Red", "Green", "Purple", "Orange", "Pink", "White"],
    "Price(N)": [250, 240, 190, 160, 200, 300, 260, 275]
}
df1 = pd.DataFrame(a)

# Creating another DataFrame Object
b = {
    "S/N": [1,2,3,4,5,6],
    "Animals": ["Goat", "Dog", "Cat", "Hen", "Duck", "Parrot"],
    "Weight(Kg)": [25, 36, 2, 4, 2.5, 1.5]
}
df2 = pd.DataFrame(b)
```

```python
df1
```

```python
df2
```

```python
# Joining along the rows
pd.concat([df1, df2], ignore_index=True)
```

```python
# Joining along the column
pd.concat(
    [df1, df2],
    ignore_index=False,
    axis=1,
    join="inner"
)
```

```python
# Adding keys to demarcate one DataFrame from the other
pd.concat(
    [df1, df2],
    ignore_index=False,
    join="outer",
    axis=0,
    keys=["Set1", "Set2"]
)
```

```python
# Appending one DataFrame to the Other
df1.append(df2)
```

#### Merging a DataFrame

```python
merged = df1.merge(df2, how="inner", on="S/N")
merged
```

### Retrieving Data From Microsoft SQL Server

```python

import pandas as pd
import sqlite3  # You can use the appropriate database connector for your database system

# Create a connection to your database
connection = sqlite3.connect('your_database.db')  # Replace with your database connection details

# Write your SQL query
sql_query = "SELECT * FROM your_table"

# Use pd.read_sql to execute the query and load the results into a DataFrame
df = pd.read_sql(sql_query, connection)

# Close the database connection when you're done
connection.close()

# Now, you can work with the data in the DataFrame
print(df)
```

## Creating a Date Range
Syntax: `pd.data_range(date,period,frequency)`
- The first parameter is the starting date
- The second parameter is the number of periods (optional if the end date is specified)
- The last parameter is the frequency: day: 'D,' month: 'M' and year: 'Y.'

```python
myDate = pd.date_range(
    start="2023-10-01",
    periods=12,
    freq='M'
)

myDate
```
