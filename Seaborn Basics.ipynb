### BAR PLOT

Create a basic BARPLOT

Add HUE to the barplot

```python
import seaborn as sb
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
a = pd.read_csv("C:\\Users\\DANIEL C E\\Desktop\\My Python Work\\Files\\SampleSuperstore.csv")
a.head()
```

```python
sb.barplot(x=a.Region,y=a["Sales"], data=a)
```

```python
sb.barplot(x=a.Region,y=a["Sales"], 
           data=a,
          hue=a.Segment)
```

Make the barplot HORIZONTAL

```python
sb.barplot(y=a.Region,x=a["Sales"], 
           data=a,
          hue=a.Segment)
```

Plot the bars in a given ORDER

```python
aa = a["Region"].unique()
aa.sort()
print(aa.tolist())
```

```python
sb.barplot(y=a.Region,x=a["Sales"], 
           data=a,
          hue=a.Segment,
          order=aa)

plt.show()
```

Add CAPP on the ERROR BAR

```python
sb.barplot(x=a.Region,y=a["Sales"], 
           data=a,
          hue=a.Segment,
          order=aa,
          capsize=0.1)

plt.show()
```

```python
sb.barplot(x=a.Region,y=a["Sales"], 
           data=a,
          hue=a.Segment,
          order=aa,
          capsize=0.05)

plt.legend(loc=(1,.8))

plt.show()
```

REMOVE the error bar using ci

```python
sb.barplot(x=a.Region,y=a["Sales"], 
           data=a,
          hue=a.Segment,
          order=aa,
          ci=None)

plt.legend(loc="center")

plt.show()
```

Change bar colors using COLOR attribute

```python
sb.barplot(x=a.Region,y=a["Sales"], 
           data=a,
          hue=a.Segment,
          order=aa,
          ci=None,
          palette="gist_heat")

plt.legend(loc="center")

plt.show()
```

Using SATURATION parameter

```python
sb.barplot(x=a.Region,y=a["Sales"], 
           data=a,
          hue=a.Segment,
          order=aa,
          ci=None,
          palette="gist_heat",
          saturation=4)

plt.legend(loc="center")

plt.show()
```

Change Default aggregation method using ESTIMATOR parameter

```python
sb.barplot(x=a.Region,y=a["Sales"], 
           data=a,
          hue=a.Segment,
          order=aa,
          ci=None,
          palette="gist_heat",
          saturation=4,
          estimator=np.sum)

plt.legend(loc="center")
#plt.grid(axis="y",which="minor")
plt.show()
```

## BOX PLOTS

Import dataset

```python
a = pd.read_csv("C:\\Users\\DANIEL C E\\Desktop\\My Python Work\\Files\\SampleSuperstore.csv")
a.head()
```

Create a BOXPLOT on a NUMERIC variable

```python
a["Category"].unique()
```

```python
sb.boxplot(data=a, y=a["Quantity"], width=.15)
```

Create a basic BOX plot on one numeric variable by a CATEGORICAL variable

```python
sb.boxplot(data=a, y=a["Quantity"], width=.5,
          x=a["Category"])
```

Create a basic BOX plot on one numeric variable by TWO CATEGORICAL variable using HUE attribute

```python
sb.boxplot(data=a, y=a["Quantity"], width=.5,
          x=a["Category"],
          hue=a.Segment)
plt.show()
```

Add MEAN marker in the box plot using SHOWMEANS attribute and change its style using meanprops

```python
sb.boxplot(data=a, y=a["Quantity"], width=.5,
          x=a["Category"],
          hue=a.Segment,
          showmeans=True,
          meanprops=dict(marker="o", markerfacecolor="w", markersize="5", mec="k"))

plt.show()
```

Make Horizontal box plot

```python
sb.boxplot(data=a, x=a["Quantity"], width=.5,
          y=a["Category"],
          hue=a.Segment,
          showmeans=True,
          meanprops=dict(marker="o", markerfacecolor="w", markersize="5", mec="k"))

plt.show()
```

Change PALETTE, LINE WIDTH etc

```python
sb.boxplot(data=a, y=a["Quantity"], width=.5,
          x=a["Category"],
          hue=a.Segment,
          showmeans=True,
          meanprops=dict(marker="o", markerfacecolor="w", markersize="5", mec="k"),
          palette="spring",
          linewidth=4)

plt.show()
```

Creating box plot for EACH OF THE NUMERIC VARIABLE in the dataframe

```python
sb.boxplot(data=a, 
           width=.5,
          showmeans=True,
          meanprops=dict(marker="o", markerfacecolor="w", markersize="5", mec="k"),
          palette="spring",
          linewidth=4)

plt.show()
```
