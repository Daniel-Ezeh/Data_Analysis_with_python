## INTRODUCTION TO SEABORN

Seaborn is a statistical data visualization library built on top of Matplotlib. It provides a high-level interface for creating attractive and informative statistical graphics. Seaborn comes with several built-in themes and color palettes to make it easy to create aesthetically pleasing visualizations. It is particularly well-suited for visualizing complex datasets with multiple variables.

#### Key features of Seaborn include:

- **High-Level Interface:** Seaborn simplifies the process of creating complex statistical visualizations. It has functions for common plot types such as scatter plots, bar plots, box plots, and more.

- **Statistical Estimation:** Seaborn includes functions for plotting statistical relationships and estimating trends in the data. For example, it provides the lmplot function for plotting linear regression models.

- **Color Palettes and Themes:** Seaborn makes it easy to change the color palette and overall look of your plots with built-in themes and color palettes. This allows for better customization and ensures that your visualizations are visually appealing.

- **Dataset-oriented API:** Seaborn is designed to work seamlessly with pandas DataFrames. You can pass a DataFrame directly to most Seaborn functions, making it convenient for working with real-world datasets.

- **Matrix Plots:** Seaborn provides functions for visualizing matrices of data, such as heatmaps, cluster maps, and pair plots. These are particularly useful for exploring relationships in multivariate datasets.

### Countplot

Syntax : seaborn.countplot(x=None, y=None, hue=None, data=None, order=None, hue_order=None, orient=None, color=None, palette=None, saturation=0.75, dodge=True, ax=None, **kwargs)
Parameters : This method is accepting the following parameters that are described below: 
 

x, y: This parameter take names of variables in data or vector data, optional, Inputs for plotting long-form data.
hue : (optional) This parameter take column name for colour encoding.
data : (optional) This parameter take DataFrame, array, or list of arrays, Dataset for plotting. If x and y are absent, this is interpreted as wide-form. Otherwise it is expected to be long-form.
order, hue_order : (optional) This parameter take lists of strings. Order to plot the categorical levels in, otherwise the levels are inferred from the data objects.
orient : (optional)This parameter take “v” | “h”, Orientation of the plot (vertical or horizontal). This is usually inferred from the dtype of the input variables but can be used to specify when the “categorical” variable is a numeric or when plotting wide-form data.

color : (optional) This parameter take matplotlib color, Color for all of the elements, or seed for a gradient palette.

palette : (optional) This parameter take palette name, list, or dict, Colors to use for the different levels of the hue variable. Should be something that can be interpreted by color_palette(), or a dictionary mapping hue levels to matplotlib colors.
saturation : (optional) This parameter take float value, Proportion of the original saturation to draw colors at. Large patches often look better with slightly desaturated colors, but set this to 1 if you want the plot colors to perfectly match the input color spec.
dodge : (optional) This parameter take bool value, When hue nesting is used, whether elements should be shifted along the categorical axis.
ax : (optional) This parameter take matplotlib Axes, Axes object to draw the plot onto, otherwise uses the current Axes.
kwargs : This parameter take key, value mappings, Other keyword arguments are passed through to matplotlib.axes.Axes.bar().

## Boxplots
Box Plot is the visual representation of the depicting groups of numerical data through their quartiles. Boxplot is also used for detect the outlier in data set. 

It captures the summary of the data efficiently with a simple box and whiskers and allows us to compare easily across groups. Boxplot summarizes a sample data using 25th, 50th and 75th percentiles. These percentiles are also known as the lower quartile, median and upper quartile.

A box plot consist of 5 things.

1. Minimum
2. First Quartile or 25%
3. Median (Second Quartile) or 50%
4. Third Quartile or 75%
5. Maximum


Syntax: `seaborn.boxplot(x=None, y=None, hue=None, data=None, order=None, hue_order=None, orient=None, color=None, palette=None, saturation=0.75, width=0.8, dodge=True, fliersize=5, linewidth=None, whis=1.5, notch=False, ax=None, **kwargs)`

- hue = feature of dataset

## Violinplots

A violin plot is a data visualization that combines aspects of a box plot and a kernel density plot. It is particularly useful for visualizing the distribution and summary statistics of a dataset. 

A violin plot plays a similar activity that is pursued through whisker or box plot do. As it shows several quantitative data across one or more categorical variables. It can be an effective and attractive way to show multiple data at several units.

Syntax: `seaborn.violinplot(x=None, y=None, hue=None, data=None, order=None, hue_order=None, bw=’scott’, cut=2, scale=’area’, scale_hue=True, gridsize=100, width=0.8, inner=’box’, split=False, dodge=True, orient=None, linewidth=None, color=None, palette=None, saturation=0.75, ax=None, **kwargs)`
Parameters: 
- x, y, hue: Inputs for plotting long-form data. 
- data: Dataset for plotting. 
- scale: The method used to scale the width of each violin. 


A violin plot is a data visualization that combines aspects of a box plot with a kernel density plot. It provides insights into the distribution of the data and is particularly useful for comparing the distribution of different groups or categories. Here's an overview of the key characteristics and components of a violin plot:

#### Key Components:
**Violin Shape:**

The main body of the violin plot represents the probability density of the data at different values. It is constructed by mirroring kernel density plots on each side.

**Box Plot Inside the Violin:**

A box plot is often included within the violin plot. The box represents the interquartile range (IQR), with a line inside indicating the median. The "whiskers" extend from the box to the maximum and minimum values within a certain range.
Width Variation:

The width of the violin plot varies along the x-axis, indicating the density of data points at different values. Wider sections represent higher density.
#### Key Concepts:
**Kernel Density Estimation (KDE):**

The kernel density plot within the violin is an estimate of the probability density function of the data. It provides a smooth representation of the distribution.
**Symmetry:**

Violin plots are often symmetrical, with mirrored distributions on either side. However, asymmetrical violins can indicate uneven data distributions.
**Data Points:**

Individual data points or a summary of data points (such as a strip plot) can be overlaid on the violin to show the actual data distribution.
#### Use Cases:
- Comparison of Distributions:

Violin plots are useful for comparing the distribution of a quantitative variable across different categories or groups.
Identification of Modes:

Multiple peaks or modes in the violin plot can indicate subgroups or clusters within the data.
- Density Estimation:

Violin plots provide a visual representation of the data density, helping to identify regions with higher or lower concentration of data.

## Strip Plot

A strip plot is a graphical data analysis technique for summarizing a univariate data set.

A strip plot is a type of categorical scatter plot used in data visualization to display individual data points along with some indication of the underlying distribution. 

    Syntax:
`seaborn.stripplot(*, x=None, y=None, hue=None, data=None, order=None, hue_order=None, jitter=True, dodge=False, orient=None, color=None, palette=None, size=5, edgecolor=’gray’, linewidth=0, ax=None, **kwargs)`

#### Parameters
    x, y, hue: Inputs for plotting long-form data.
    data: Dataset for plotting.
    order: It is the order to plot the categorical levels in.
    color: It is the color for all of the elements, or seed for a gradient palette    
    marker: The marker style for the points (e.g., 'o' for circles, 's' for squares).

#### Additional Options:
    hue: Grouping variable that will produce points with different colors. This is useful for visualizing an additional categorical variable.

    order, hue_order: Specify the order of the categories on the plot.

    size: Grouping variable that will produce points with different sizes.

    dodge: When hue nesting is used, setting dodge to True will separate the strips for different hues.

#### Interpretation:
**Point Distribution:** Each point represents an individual data point. The vertical position of points on the plot provides information about their distribution along the categorical axis.

**Density:** The concentration of points at specific positions gives an indication of the density of the data.

**Categorical Comparisons:** Strip plots are useful for comparing the distribution of a variable across different categories.

## Swarm Plot
A swarm plot is a type of categorical scatter plot used in data visualization to show individual data points along with the distribution of the data. It is particularly useful when you have a relatively small dataset and want to display each data point

#### Key Components:
**Individual Data Points:**

    Each data point in the dataset is represented by a marker on the plot, allowing for a detailed view of the distribution.
**Avoidance of Overlapping:**

    Swarm plots use an algorithm to position individual points without overlapping. This helps prevent the crowding of points and allows for a clearer visualization of the data distribution.
#### Key Concepts:
**Categorical Variables:**

    Swarm plots are typically used to visualize the distribution of a quantitative variable across different categories.
**Point Density:**

    The density of points at specific positions provides insights into the distribution of the data.
#### Use Cases:
**Comparison of Distributions:**

    Swarm plots are useful for comparing the distribution of a quantitative variable across different categories or groups.
**Identification of Individual Data Points:**

    Swarm plots are effective when you want to see the distribution of each individual data point.
