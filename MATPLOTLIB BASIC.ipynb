Syntax: matplotlib.pyplot.figure.add_axes(rect)

rect: This parameter is the dimensions `[xmin, ymin, dx, dy]` of the new axes. It takes the below elements as arguments in the list:

- `xmin`: Horizontal coordinate of the lower left corner.
- `ymin`: Vertical coordinate of the lower left corner.
- `dx`: Width of the subplot.
- `dy`: Height of the subplot.

Returns: This method return the axes class depends on the projection used. 

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt


x = np.arange(0, 12, 0.05)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = y1+y2

# Task 1: Changing the color of the plot to a red line and customizing other properties

# Creating a new figure with width = 6 inches
# and height = 5

fig = plt.figure(figsize=(3,2), dpi=500, frameon=True, facecolor="m")

# Creating first axes for the figure
ax1 = fig.add_axes([0,0,0.8,0.5])
ax1.set_facecolor("black")
ax1.grid(which="both", axis='y')
ax1.plot(x, y1, c="r", lw=5)

# Creating second axes for the figure
ax2 = fig.add_axes([0,0.55,0.8,0.5])
ax2.plot(x, y2, ":k")

# Creating third axes for the figure
ax3 = fig.add_axes([0.8,0,0.3,1.05])
ax3.plot(x, y3, alpha=0.2)



plt.show()
```

### SUB PLOTS

```python
# Creating 2 subplot in a plot 

plt.subplot(1,3,1) # This is the first subplot
plt.plot(x, y1, c="r", lw=5)

p2 = plt.subplot(1,3,2) # This is the second subplot
p2.set_facecolor("#87F5EB")
p2.plot(x, y2, ":k")

p3 = plt.subplot(1,3,3) # This is the third subplot
p3.plot(x, y3, alpha=0.2)

plt.subplots_adjust(wspace=.4)

plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt


x = np.arange(0, 12, 0.05)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = y1+y2


# Bar
x1 = np.arange(1,7)
col = ["Blue", "Red", "Green", "Yellow", "Pink", "Purple"]
y = np.array([127.08, 26.0, 148.0, 100.0, 10.69, 127.5])
plt.figure(figsize=(10,7))

p1 = plt.subplot(2,3,1) # This is the first subplot
p1.plot(x, y1, c="r", lw=5)

p2 = plt.subplot(2,3,2) # This is the second subplot
p2.set_facecolor("#87F5EB")
p2.plot(x, y2, ":k")

p3 = plt.subplot(2,3,3) # This is the third subplot
p3.plot(x, y3, alpha=0.2)

p4 = plt.subplot(2,3,4) # This is the fourth subplot
p4.pie(y[:3], 
        labels=col[:3], 
        explode=[0,0.2,0], 
        autopct="%.2f%%")

p5 = plt.subplot(2,3,5) # This is the fifth subplot
p5.bar(col, y, color=col)

p6 = plt.subplot(2,3,6) # This is the sixth subplot
p6. plot(y, marker="+", mec="red", ms=10)

plt.subplots_adjust(wspace=.2, hspace=.2)
plt.show()
```

#### Text and Annotation

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

y1 = np.array([127.08,  26.,   148.,   100.,    10.69, 127.5 ])
y2 = np.array([301.32,  28.,   479.,   119.,    73.62, 291.2 ])

fd = dict(family="cursive", fontsize="x-large", c="black", style="italic")

plt.figure(figsize=(10,5))

line1 = plt.plot(y1, "D--r", ms=5, mec="black", mfc="darkblue", alpha=0.3)
line2 = plt.plot(y2, "*-k", ms=8, mec="red", mfc="hotpink")

plt.text(0,400, "This is a graph that shows \n The Pulse and MaxPulse", fontdict=fd)

#plt.grid()

# Adding an annotation
plt.annotate("This is the maximum point", xy=(2,480), xytext=(3.5,400),
            arrowprops=dict(facecolor="red", shrink=0.1)
            )

plt.show()
```

### Creating a 3D plot

```python
# Importing the necessary libraries 

import mpl_toolkits.mplot3d as dd # the same as
from mpl_toolkits import mplot3d as dd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import pyplot as plt

plt.style.use("seaborn")

z = np.linspace(0,1,200)
x = z*np.sin(25*z)
y = z*np.cos(25*z)

fig = plt.figure(figsize=(10,5))
ax1 = plt.axes([0,0,1,1], projection="3d")
ax1.plot3D(x,y,z)

ax2 = plt.axes([1,0,0.5,1])
ax2.plot(z,y)

ax3 = plt.axes([1.5,0,0.5,1])
ax3.plot(z,x, c="r")

fig.savefig("C:\\Users\\DANIEL C E\\Desktop\\My Python Work\\Files\\Plots\\fig4.jpeg")
plt.show()
```
